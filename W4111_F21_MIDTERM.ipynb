### Birth Countries and Death Countries

<u>Question</u>

- In ```lahmansbaseballdb.people``` there is information about people's ```birthCountry``` and ```deathCountry.```



- There are countries in which at least person was born but in which no person has died. 


- Write a query that produces a table of the form:
    - ```birthCountry```
    - ```no_of_births```, which is the total number of births in the country
    
    
- The table contains all rows in which there with births but no deaths.

<u>Answer</u>


````python
%%sql
select birthCountry, count(playerID) as no_of_births
from people
where birthCountry not in (select ifnull(deathCountry,'')  from people)
group by birthCountry
order by no_of_births desc;
````

### Best Baseball Players

<u>Question</u>

- This question uses ```lahmansbaseballdb.batting```, ```lahmansbaseballdb.pitching``` and ```lahmansbaseballdb.people```.


- There query computes performance metrics:
    - Batting:
        - On-base percentage: OBP is (sum(h) + sum(BB))/(sum(ab) + sum(BB))
        - Slugging percentage: SLG is        
```
    (
            (sum(h) - sum(`2b`) - sum(`3b`) - sum(hr)) +
            2*sum(`2b`) + 3*sum(`3b`) + 4*sum(hr)
            )/sum(ab)
```
        - On-base percentage plus slugging: OPS is is ```(obp + slg)```.
    - Pitching:
        - total_wins is ```sum(w)```.
        - total_loses is ```sum(l)```.
        - win_percentage is ```sum(w)/(sum(w) + sum(l))```.
        
        
- Professor Ferguson has two criteria for someone being a great baseball player.
    - Batting:
        - Total number of ```ab >= 1000```.
        - OPS: Career ```OPS >= 1.000```
    - Pitching:
        - ```(sum(w) + sum(l)) >= 200```.
        - ```win_percentage >= 0.70)``` or ```sum(w) >= 300.```
        
        
- This is one of the rare cases where Prof. Ferguson will provide the answer. So, please produce the table below. Some notes:
    - ```great_because``` is either ```Pitcher``` or ```Batter``` based on whether the player matched the batting or pitching criteria.
    - The values from ```batting``` are ```None``` if the player did not qualify based on batting.
    - The values from ```pitching``` are ```None``` if the player did not qualify on pitching.


<u>Answer</u>


````python
%%sql
with b as
    (
        select playerID,sum(ab) as total_ab,
        (sum(h) + sum(BB))/(sum(ab) + sum(BB)) as OBP,
        ((sum(h) - sum(`2b`) - sum(`3b`) - sum(hr)) +2*sum(`2b`) + 3*sum(`3b`) + 4*sum(hr))/sum(ab) as SLG,
        ((sum(h) + sum(BB))/(sum(ab) + sum(BB)))+(((sum(h) - sum(`2b`) - sum(`3b`) - sum(hr)) +2*sum(`2b`) + 3*sum(`3b`) + 4*sum(hr))/sum(ab)) as OPS
        from batting
        group by playerID
    ),
  p as (
      select playerID,sum(w) as total_wins,sum(L) as total_loses,sum(w)/(sum(w) + sum(l)) as win_percent
      from pitching
      group by playerID
  ),
  great_b as (
      select playerID, total_ab, OBP, SLG, OPS
      from b
      where total_ab > 1000 and OPS > 1.0
  ),
  great_p as (
      select playerID, total_wins, total_loses, win_percent
      from p
      where (total_wins+total_loses) >= 200 and (win_percent >= 0.70 or total_wins >= 300)
  )
    select great_b.playerID,p.nameLast,p.nameFirst,'Batter' as great_because,p.debut_date,p.finalgame_date,OBP,SLG,OPS,total_wins,total_loses,win_percent
    from great_b left join great_p using(playerID) join people p using(playerID)
    union
    select great_p.playerID,p.nameLast,p.nameFirst,'Pitcher' as great_because,p.debut_date,p.finalgame_date,OBP,SLG,OPS,total_wins,total_loses,win_percent
    from great_p left join great_b using(playerID)join people p using(playerID);
````

````python
%%sql
drop view if exists Person;
create view Person as
    select id, name, dept_name, 'I' as role
    from instructor
    union
    select id, name, dept_name, 'S' as role
    from student;
select * from Person;
````

_Create procedure statement_

````python
%%sql
drop procedure if exists create_person;
CREATE DEFINER=`root`@`localhost` PROCEDURE `create_person`(
    in person_name varchar(32),
    in dept_name varchar(32),
    in total_cred decimal(3,0),
    in salary decimal(8,2),
    out out_ID varchar(5)
    )
BEGIN

    declare bad_person boolean;
    declare new_id varchar(12);

    set bad_person = false;
    set new_id = '00000';


    /*
        The logic of the stored procedure is the following:
            - The request is invalid and sets bad_person to true if:
                - Any of person_name, dept_name is NULL.
                - Either total_cred is NULL and salary is NOT NULL, or salary is NULL and total_cred is NULL.
            - The procedure must compute a new, unique ID. The approach is to find the manimum
                ID value over Student and Instructor. Add 1 to the value to produce the new, unique ID.
            - The procedure then adds the information to Student or Instructor based on whether
                total_cred is NULL or salary is NULL.
    */
    if person_name is null or dept_name is null then
        set bad_person = true;
    end if;

    if (salary is null and total_cred is null) or (salary is not null and total_cred is not null) then
        set bad_person = true;
    end if;

    if bad_person is false then
        select max(ID)+1 into out_ID from (select ID from student union select ID from instructor)as t;
        if total_cred is null then
            insert into instructor values (out_ID, person_name, dept_name, salary);
        end if;
        if salary is null then
            insert into student values (out_ID, person_name, dept_name, tot_cred);
        end if;
    end if;

    if bad_person is true then
        SIGNAL SQLSTATE '50001'
            SET MESSAGE_TEXT = 'Invalid Person information input';
    end if;

    /* NOTE: ID is the out parameter. You must set ID to the new, unique ID */

END;
select * from Person;
````

````python
%%sql
alter table worldcities modify city varchar(50) null;

alter table worldcities modify city_ascii varchar(50) null;

alter table worldcities modify country varchar(50) null;

alter table worldcities modify iso2 varchar(50) null;

alter table worldcities modify iso3 varchar(50) null;

create table countries
(
	country varchar(50) not null,
	iso2 varchar(50) not null,
	iso3 varchar(50) not null,
	constraint counties_pk
		primary key (iso2)
);

create unique index counties_country_uindex
	on countries (country);

create unique index counties_iso3_uindex
	on countries (iso3);

create table cities
(
	city varchar(50) not null,
	city_ascii varchar(50) not null,
	lat double not null,
	lng double not null,
	iso3 varchar(50) not null,
	id bigint not null,
	constraint cities_pk
		primary key (city),
	constraint cities_countries_iso3_fk
		foreign key (iso3) references countries (iso3)
);

create unique index cities_city_ascii_uindex
	on cities (city_ascii);

create unique index cities_id_uindex
	on cities (id);

create unique index cities_iso3_uindex
	on cities (iso3);
    
select * from countries;
````
